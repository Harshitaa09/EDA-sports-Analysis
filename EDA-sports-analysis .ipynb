```python
plt.figure(figsize = (10,6))
sns.countplot('season',data=data,palette="Oranges")
plt.title("Number of Matches played in each IPL season",fontsize=20)
plt.xlabel("season",fontsize=17)
plt.ylabel('Matches',fontsize=17)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.countplot(x='winner',data=data, palette='Oranges_r')
plt.title("Numbers of matches won by team",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel("Teams",fontsize=16)
plt.ylabel("No of wins",fontsize=16)
plt.show()
```

```python
data['win_by']=np.where(data['win_by_runs']>0,'Bat first','Bowl first')
```

```python
Win=data.win_by.value_counts()
Win.head()
```

<b><font size="3">If in the data win_by_runs>0 then the team choose to bat first otherwise choose to bowl first </font></b>

```python
Win=data.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
colors = ['#F4FC5E', '#FCB125']
plt.figure(figsize = (7,5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%5.1f%%',startangle=0)
plt.title('Match Result',fontsize=20)
plt.show()
```

```python
plt.figure(figsize = (10,7))
sns.countplot('season',hue='win_by',data=data,palette='coolwarm')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=17)
plt.ylabel("Count",fontsize=17)
plt.show()
```

<b><font size="3">Teams that opt for bowling first tend to win more than the ones who opt for batting</font></b>

```python
# we will plot pie chart on Toss decision
Toss=data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#48cefa', '#b06bed']
plt.figure(figsize = (7,5))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%7.1f%%', shadow=True,startangle=0)
plt.title('Toss result',fontsize=20)
plt.axis('equal')
plt.show()
```

<b><font size="3">This pie chart tells that the toss winner opt to bat or bowl , so 61.2% winners opt to field,
since toss_decisions only consist of two options that is to bat or bowl so no need to put np.where condition</font></b>

```python
#printing numbers of fours hit by team
four_data=complete_data[complete_data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python

#ploting graph on four hit by players
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10].plot('batsman','four',kind='bar',color='#f5b73b')
plt.title("Numbers of 4's hit by playes ",fontsize=20)
plt.xticks(rotation=30)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

```python
#ploting graph on no of four hit in each season
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color = '#f0d122')
plt.title("Numbers of 4's hit in each season ",fontsize=20)
plt.xticks(rotation=30)
plt.xlabel("season",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

```python
#printing no of sixes hit by team
six_data=complete_data[complete_data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
#ploting graph of six hit by players
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='#ad5afa')
plt.title("Numbers of 6's hit by playes ",fontsize=20)
plt.xticks(rotation=30)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of six",fontsize=15)
plt.show()
```

```python
#ploting graph on no of six hit in each season
ax=six_data.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color = '#3cc9f0')
plt.title("Numbers of 6's hit in each season ",fontsize=20)
plt.xticks(rotation=30)
plt.xlabel("season",fontsize=15)
plt.ylabel("No of six",fontsize=15)
plt.show()
```

```python
#printing the top 10 leading run scorer in IPL
batsman_score=Data.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("###TOP 10 LEADING RUN SCORER IN IPL###")
batsman_score.iloc[:10,:]
```

```python
wicket_data=Data.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
wicket_data.head(2)
```
